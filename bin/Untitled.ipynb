```python
from pint import UnitRegistry
import math
```

```python


# Инициализация UnitRegistry
ureg = UnitRegistry()
Q_ = ureg.Quantity

# === Входные данные (измени под свои значения) ===
mass_flow = Q_(0.3, 'kg/s')        # массовый расход
radius = Q_(25, 'mm')            # радиус сечения
density = Q_(2700, 'kg/m**3')     # плотность жидкости (например, вода при 4°C)

# === Расчёт скорости ===
area = math.pi * radius**2                # площадь поперечного сечения
velocity = mass_flow / (density * area)   # скорость = расход / (плотность * площадь)

# Приведение к метрам в секунду
velocity = velocity.to('m/s')

print(f"Скорость потока: {velocity:.3f}")
```
